```python
avukat_yas = {
    'Genc': ["Sanığın yaşı ({failYasi}) küçüktür, önünde uzun bir gelecek vardır."],
    'Orta': ["Müvekkil ({failYasi}) yaşında, iş ve aile düzeni olan biridir."],
    'Yasli': ["Müvekkil ({failYasi}) yaşındadır, sağlık durumu cezaevi koşullarına uygun değildir."]
}
```

```python
def metin_uretici_test(row):
    tutum = row['sanikTutumu']
    strateji = row['avukatStratejisi']
    yas = int(row['failYasi'])
    deger = int(row['malinDegeriTL'])
    vakit = str(row['sucVakti'])
    kilit = str(row['kilitDurumu'])

    # Tutarlilik icin
    if tutum in ['Inkar', 'Agresif', 'Celiski']:
        if strateji == 'IndirimOdakli':
            strateji = 'DelilYetersizligi'

    # Pisman ise delil yetersizligi sacma olurdu
    if tutum == 'Pisman':
        if strateji == 'DelilYetersizligi':
            strateji = 'IndirimOdakli'

    # Deger yuksekken defer azligi savunmasi yapamaz
    if strateji == 'VasifDegisikligi':
        if deger < 3000:
            if random.random() < 0.5:
                final_strateji = 'VasifDegisikligi_DegerAzligi'
            else:
                final_strateji = 'VasifDegisikligi_Kullanma'
        else:
            final_strateji = 'VasifDegisikligi_Kullanma'

            if tutum == 'Inkar':
                final_strateji = 'DelilYetersizligi'
                strateji = 'DelilYetersizligi'
    else:
        final_strateji = strateji

    # Sanik metni
    if kilit == 'KilitKirilarak': kilit_txt = "kilidi kırarak"
    else: kilit_txt = "kapıyı zorlamadan"

    if yas < 18: hitap = random.choice(sanik_hitaplar['Cocuk'])
    elif tutum in ['Agresif', 'Inkar']:
        hitap = random.choice(sanik_hitaplar['Kaba']) if random.random() < 0.7 else random.choice(sanik_hitaplar['Genel'])
    else: hitap = random.choice(sanik_hitaplar['Genel'])

    giris = random.choice(sanik_giris.get(tutum, sanik_giris['Inkar']))
    gelisme = "" if tutum == 'Sessiz' else random.choice(sanik_gelisme.get(tutum, sanik_gelisme['Inkar']))
    sonuc = random.choice(sanik_sonuc.get(tutum, sanik_sonuc['Inkar']))

    if tutum == 'Sessiz': sanik_text = giris
    elif random.random() < 0.5: sanik_text = f"{hitap} {giris} {gelisme} {sonuc}"
    else: sanik_text = f"{hitap} {sonuc} {gelisme}"

    sanik_text = sanik_text.strip().replace(" ,", ",").replace("  ", " ")
    if sanik_text.startswith(","): sanik_text = sanik_text[1:].strip()

    try:
        sanik_text = sanik_text.format(sucVakti=vakit.lower(), sucMekani=row['sucMekani'], malinDegeri=deger, kilitDurumu=kilit_txt)
    except: pass

    # --- 3. AVUKAT METNİ ---
    aktif_havuz = avukat_argumanlar.get(final_strateji, avukat_argumanlar['DelilYetersizligi'])[:]

    if strateji == 'IndirimOdakli' or random.random() < 0.2:
        if yas < 25: aktif_havuz.extend(avukat_yas['Genc'])
        elif yas > 60: aktif_havuz.extend(avukat_yas['Yasli'])
        else: aktif_havuz.extend(avukat_yas['Orta'])

    secilenler = random.sample(aktif_havuz, k=min(len(aktif_havuz), random.randint(1, 3)))
    avukat_text = f"{random.choice(avukat_giris)} {' '.join(secilenler)} {random.choice(['Beraat talep ediyoruz.', 'Takdir mahkemenindir.'])}"

    try:
        avukat_text = avukat_text.format(failYasi=yas, malinDegeri=deger, sucVakti=vakit.lower(), kilitDurumu=kilit_txt)
    except: pass

    return sanik_text, avukat_text, strateji
```

```python
if not dfFinal.empty:

    temp_res = dfFinal.apply(lambda row: metin_uretici_test(row), axis=1)

    dfFinal['sanik_ifadesi'] = temp_res.apply(lambda x: x[0])
    dfFinal['avukat_savunmasi'] = temp_res.apply(lambda x: x[1])
    dfFinal['avukatStratejisi'] = temp_res.apply(lambda x: x[2])

    # NLP Çıktısı
    df_nlp_test = dfFinal[[
        'sanik_ifadesi', 'avukat_savunmasi',
        'sanikTutumu', 'avukatStratejisi'
    ]].copy()

    print("Sanik davranis dagilimi")
    print(df_nlp_test['sanikTutumu'].value_counts())
    print()

    print("Avukat strateji dagilimi")
    print(df_nlp_test['avukatStratejisi'].value_counts())
    print()

    df_nlp_test.to_csv(CIKIS_DOSYASI, index=False)
    print(f"Dosya kaydedildi")

    try:
        files.download(CIKIS_DOSYASI)
    except:
        pass
```
